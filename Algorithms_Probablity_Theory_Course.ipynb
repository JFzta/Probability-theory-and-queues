```python
import numpy as np
import math
import itertools


# Datos de la tabla de frecuencia
clases = [(155, 161), (161, 167), (167, 173), (173, 179), (179, 185), (185, 191)]
frecuencia = [2, 5, 5, 7, 6, 3]

# Calcular la marca de clase (punto medio de cada intervalo)
marcas_clase = [(a + b) / 2 for a, b in clases]

# Calcular la media
media = sum(m * f for m, f in zip(marcas_clase, frecuencia)) / sum(frecuencia)

# Calcular la moda (intervalo con mayor frecuencia)
moda_intervalo = clases[frecuencia.index(max(frecuencia))]

# Calcular la mediana
N = sum(frecuencia)
frecuencia_acumulada = np.cumsum(frecuencia)
mediana_intervalo = None
for i, Fi in enumerate(frecuencia_acumulada):
    if Fi >= N / 2:
        mediana_intervalo = clases[i]
        break

# Método para calcular percentil usando la fórmula para datos agrupados
def calcular_percentil(p):
    """Calcula el percentil p usando la fórmula para datos agrupados."""
    posicion = (p / 100) * N
    for i, Fi in enumerate(frecuencia_acumulada):
        if Fi >= posicion:
            LI = clases[i][0]  # Límite inferior del intervalo
            Fi_ant = frecuencia_acumulada[i - 1] if i > 0 else 0  # Frecuencia acumulada anterior
            fi = frecuencia[i]  # Frecuencia del intervalo
            A = clases[i][1] - clases[i][0]  # Amplitud del intervalo
            percentil = LI + ((posicion - Fi_ant) / fi) * A
            return percentil
    return None

def calcular_varianza(muestral=False):
    """Calcula la varianza de los datos agrupados."""
    suma_cuadrados = sum(f * (m - media) ** 2 for m, f in zip(marcas_clase, frecuencia))
    if muestral:
        return suma_cuadrados / (N - 1)  # Varianza muestral
    return suma_cuadrados / N  # Varianza poblacional


# Mostrar tabla de frecuencias
print("Tabla de Frecuencias")
print("| Intervalo | fi | Marca de Clase |")
print("|-----------|----|---------------|")
for i in range(len(clases)):
    print(f"| {clases[i]} | {frecuencia[i]} | {marcas_clase[i]:.2f} |")

# Mostrar resultados
print(f"\nMedia: {media:.4f}")
print(f"Moda: Intervalo {moda_intervalo}")
print(f"Mediana: Intervalo {mediana_intervalo}")

# Ejemplo de cálculo de percentil
percentil = calcular_percentil(30)
print(f"Percentil: {percentil:.4f}")

def calcular_permutaciones(iterable, r=None):
    """Calcula las permutaciones de un iterable dado."""
    return list(itertools.permutations(iterable, r))

varianza_poblacional = calcular_varianza()
varianza_muestral = calcular_varianza(muestral=True)
desviacion = math.sqrt(varianza_poblacional)
print(f"Varianza Poblacional: {varianza_poblacional:.10f}")
print(f"Varianza Muestral: {varianza_muestral:.5f}")
print(f"desviacion estandar: {desviacion:.5f}")

print("\nEjemplo de permutaciones de [1, 2, 3]:")
permutaciones = calcular_permutaciones([1, 2, 3])
print(permutaciones)

dias=365

def maneras_cumpleanos_distintos(n):
    """Calcula de cuántas maneras dos personas pueden cumplir años en fechas distintas."""
    return n * (n - 1)

cc= math.comb(8,4)*math.comb(6,3)

math.factorial(5)


def calcular_probabilidad_baloto_con_especial():
    n, k = 43, 5
    combinaciones = math.comb(n, k)  # Combinaciones de los 6 números
    numero_especial = 16  # Opciones para el número especial
    total_combinaciones = combinaciones * numero_especial
    probabilidad = 1 / total_combinaciones
    return probabilidad

print(f"La probabilidad de ganar el Baloto con número especial es: {calcular_probabilidad_baloto_con_especial():.12f}")



def probabilidad_placa_sin_repetidos():
    total_placas = (26**3) * (10**3)  # Todas las combinaciones posibles
    placas_sin_repetidos = (26 * 25 * 24) * (10 * 9 * 8)  # Sin repetir caracteres
    return placas_sin_repetidos / total_placas

print(f"La probabilidad de que una placa tenga todos sus elementos diferentes es: {probabilidad_placa_sin_repetidos():.4f}")
```

```python
import numpy as np
import math
import itertools

class Estadistica:
    def __init__(self, datos_puntuales=None, datos_agrupados=None):
        self.datos_puntuales = datos_puntuales
        self.datos_agrupados = datos_agrupados

        if datos_agrupados:
            self.clases, self.frecuencias = zip(*datos_agrupados)
            self.marcas_clase = [(a + b) / 2 for a, b in self.clases]
            self.N = sum(self.frecuencias)
            self.frecuencia_acumulada = np.cumsum(self.frecuencias)

    def calcular_media(self):
        if self.datos_puntuales:
            return np.mean(self.datos_puntuales)
        return sum(m * f for m, f in zip(self.marcas_clase, self.frecuencias)) / self.N

    def calcular_mediana(self):
        if self.datos_puntuales:
            return np.median(self.datos_puntuales)
        for i, Fi in enumerate(self.frecuencia_acumulada):
            if Fi >= self.N / 2:
                return self.clases[i]
        return None

    def calcular_moda(self):
        if self.datos_puntuales:
            return max(set(self.datos_puntuales), key=self.datos_puntuales.count)
        return self.clases[self.frecuencias.index(max(self.frecuencias))]

    def calcular_percentil(self, p):
        posicion = (p / 100) * self.N
        for i, Fi in enumerate(self.frecuencia_acumulada):
            if Fi >= posicion:
                LI = self.clases[i][0]
                Fi_ant = self.frecuencia_acumulada[i - 1] if i > 0 else 0
                fi = self.frecuencias[i]
                A = self.clases[i][1] - self.clases[i][0]
                return LI + ((posicion - Fi_ant) / fi) * A
        return None

    def calcular_rango_intercuartil(self):
        return self.calcular_percentil(75) - self.calcular_percentil(25)

    def calcular_varianza(self, muestral=False):
        if self.datos_puntuales:
            return np.var(self.datos_puntuales, ddof=1 if muestral else 0)
        suma_cuadrados = sum(f * (m - self.calcular_media())**2 for m, f in zip(self.marcas_clase, self.frecuencias))
        return suma_cuadrados / (self.N - 1 if muestral else self.N)

    def calcular_desviacion_estandar(self):
        return math.sqrt(self.calcular_varianza())

    def mostrar_tabla_frecuencias(self):
        if not self.datos_agrupados:
            return "No hay datos agrupados."
        print("| Intervalo | fi | Marca de Clase |")
        print("|-----------|----|---------------|")
        for i in range(len(self.clases)):
            print(f"| {self.clases[i]} | {self.frecuencias[i]} | {self.marcas_clase[i]:.2f} |")

# Ejemplo de uso
datos_agrupados = [((155, 161), 2), ((161, 167), 5), ((167, 173), 5), ((173, 179), 7), ((179, 185), 6), ((185, 191), 3)]
estadistica = Estadistica(datos_agrupados=datos_agrupados)

estadistica.mostrar_tabla_frecuencias()
print(f"Media: {estadistica.calcular_media():.4f}")
print(f"Moda: {estadistica.calcular_moda()}")
print(f"Mediana: {estadistica.calcular_mediana()}")
print(f"Percentil 30: {estadistica.calcular_percentil(30):.4f}")
print(f"Rango intercuartil: {estadistica.calcular_rango_intercuartil():.4f}")
print(f"Varianza Poblacional: {estadistica.calcular_varianza():.10f}")
print(f"Desviación Estándar: {estadistica.calcular_desviacion_estandar():.5f}")
```

```python
import math


def permutaciones(n, r):
    if r > n:
        return 0
    return math.factorial(n) // math.factorial(n - r)

# Ejemplo: cuántas formas de que 2 personas cumplan años en días distintos
resultado = permutaciones(365, 2)
print("P(365, 2) =", resultado)

def combinaciones(n, r):
    if r > n:
        return 0
    return math.factorial(n) // (math.factorial(r) * math.factorial(n - r))

# Ejemplo: combinaciones de 2 días distintos para cumpleaños
resultado_comb = combinaciones(365, 2)
print("C(365, 2) =", resultado_comb)


def permutaciones_indistinguibles(n, repeticiones):
    # Verificamos que la suma de repeticiones sea igual a n
    if sum(repeticiones) != n:
        raise ValueError("La suma de las repeticiones debe ser igual a n")

    # Numerador: n!
    numerador = math.factorial(n)

    # Denominador: producto de los factoriales de las repeticiones
    denominador = 1
    for r in repeticiones:
        denominador *= math.factorial(r)

    return numerador // denominador

# Ejemplos de uso
print(permutaciones_indistinguibles(6, [3, 2, 1]))  # BANANA → 60
print(permutaciones_indistinguibles(4, [2, 2]))     # MAMA → 6
```
